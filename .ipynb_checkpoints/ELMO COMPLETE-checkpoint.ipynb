```python
import pandas as pd
import tensorflow_hub as hub
import numpy as np
import tensorflow as tf

from tensorflow.compat.v1 import ConfigProto
from tensorflow.compat.v1 import InteractiveSession

config = ConfigProto()
config.gpu_options.allow_growth = True
sess = tf.Session(config=config)

train = pd.read_csv("train.csv")
train = train.dropna()
train[train.isnull().any(axis=1)]
```

```python
url = "https://tfhub.dev/google/elmo/2"
embed = hub.Module(url)
```

```python
# This tells the model to run through the 'sentences' list and return the default output (1024 dimension sentence vectors).
X_train = train.text.tolist()
X_train = np.array(X_train, dtype=object)
```

```python
X_train.shape
```

```python
# This tells the model to run through the 'sentences' list and return the default output (1024 dimension sentence vectors).
embeddings = embed(
    X_train,
    signature="default",
    as_dict=True)["default"]
```

```python
#Start a session and run ELMo to return the embeddings in variable x

session.run(tf.global_variables_initializer())
session.run(tf.tables_initializer())
x = session.run(embeddings)
```
